# <font color='red'>1. Happy number

Here we will see how to detect a number n is one Happy number or not. So the happy number is a number, where starting with any positive integers replace the number by the sum of squares of its digits, this process will be repeated until it becomes 1, otherwise it will loop endlessly in a cycle. Those numbers, when the 1 has found, they will be happy number.

Suppose the number is 19, the output will be true as the number is happy number. As we can see from 19, we will get


Suppose the number is 20 <br>
4 + 0 = 4<br>
So it not a happy number

The first few unhappy numbers are 2, 3, 4, 5, 6, 8, 9, 11, 12, 14, 15, 16, 17, 18, 20

```python
n = int(input("Enter a number to check weather given number is happy number or not:"))
x = n
while x>=10:
    sum1 = 0
    while x>0:
        r = x%10
        sum1 = sum1 + (r**2)
        x = x//10
    x = sum1
if x == 1:
    print(n,"is a happy number")
else:
    print(n,"is not a happy number")
    
```

# <font color='red'>2. Repete Number finder :

Input :<br>
* n = 7<br>
* list1 = [123,345,456,678,890,123,567]<br>
* k = 4
    
Output :<br>
* 890    

Explanation: 
* first input having number elements.
* then user will give the list of elements that mentioned in n time. here only on repatedt number will be there.
* then we need to provide k value. k is used to find the final answer.


Example :
* n=7 mean we will get 7 elements list.
* here we can see that 123 is repated 2 times then that 123 wil remove the remaining list is [345,456,678,890,567]
* now kth element from this list is final asnwer. So ans is 890.
    
    

```python
n = 7
list1 = [123,345,456,678,890,123,567]
k = 4
```

```python
def fun(lst):
    repeated = []
    for i in range(len(lst)):
        for j in range(i+1,len(lst)):
            if lst[i] == lst[j] and lst[i] not in repeated:
                repeated.append(lst[i])
    res = [i for i in lst if i not in repeated ]
    return res[k-1]
```

```python
fun(list1)
```
